```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import plotly.graph_objs as go

import plotly
import plotly.express as px
pd.set_option('display.max_rows', 400)
```

```python
# Обращаемся к файлу
xls=pd.ExcelFile('C:/Users/Владимир/PycharmProjects/task.xlsx')
```

```python
# Определяем листы с таблицами
df_payroll = pd.read_excel(xls, 'Payroll',converters={'ID_supplier': str})
df_payments = pd.read_excel(xls, 'Payments',converters={'ID_supplier': str})
```

```python
# Смотрим платежную ведомость
df_payroll.head()
```

```python
# Создаем id для того, чтобы сделать join с таблицей df_payments 
df_payroll['id']=df_payroll['Register_number'].astype(str)+"_"+df_payroll["Period_month"].astype(str)+"_"+ df_payroll["Period_year"].astype(str)+"_"+df_payroll["Project"].astype(str)+"_"+df_payroll["ID_supplier"].astype(str)
```

```python
# Количество уникальных значений
df_payroll.nunique()
```

```python
# Смотрим платежи
df_payments.head()
```

```python
# Строки, столбцы
df_payments.shape
```

```python
# Создаем id для того, чтобы сделать join с таблицей df_payroll
df_payments['id']=df_payments['Register_number'].astype(str)+"_"+df_payments["Period_month"].astype(str)+"_"+ df_payments["Period_year"].astype(str)+"_"+df_payments["Project"].astype(str)+"_"+df_payments["ID_supplier"].astype(str)
```

```python
# Смотрим платежи c id для операции join
df_payments.head()
```

```python
# Делаем join таблиц
df_merged=pd.merge(df_payments, df_payroll, how='right',left_on='id', right_on='id')
```

```python
# После операции join видно, что информации по срокам оплаты некоторым поставщикам нет (NaN значения). Не будем использовать их в анализе
df_merged
```

```python
# таблица без сроков оплаты
df_merged_na=df_merged.tail(12026-11396)
```

```python
df_merged_na
```

```python
# Отметим, что здесь осталось 378 уникальных ID_supplier_y, которых, скорее всего,нет в таблице Payments
df_merged_na.nunique()
```

```python
# Меняем NaN на дефолтное значение 999 для удобства 
df_merged=df_merged.fillna(999)
```

```python
df_merged['Day_payment'] = df_merged['Day_payment'].astype(np.int64)
df_merged['Month_payment'] = df_merged['Month_payment'].astype(np.int64)
df_merged['Year_payment'] = df_merged['Year_payment'].astype(np.int64)
```

```python
# Смотрим таблицу
df_merged
```

```python
# Создаем поле date
df_merged['date']=df_merged["Year_payment"].astype(str)+ '-' + df_merged["Month_payment"].astype(str) + '-' + df_merged["Day_payment"].astype(str)
df_merged['date']=pd.to_datetime(df_merged['date'],errors ='coerce')
```

```python
df_merged
```

```python
# Убираем лишние столбцы
df_cleaned=df_merged.drop(['Register_number_x', 'Period_month_x','Period_year_x','Project_x','ID_supplier_x','Month_payment','Year_payment'], axis=1)
```

```python
# Смотрим
df_cleaned
```

```python
#  Посмотрим, сколько платежей после 25 числа каждого месяца. Вероятно, это и будет просрочка по платежу
payments_not_in_time=df_cleaned[(df_cleaned['Day_payment']>25) & (df_cleaned['Day_payment']<=31)]
```

```python
payments_not_in_time
```

```python
# payments_not_in_time=table_without_na[table_without_na['Day_payment']>25]
```

```python
# payments_not_in_time.nunique()
```

```python
# Группируем данные на атомарном уровне, чтобы видеть характеристику каждого платежа 
payments_not_in_time.groupby(['Project_y','Bank','Register_number_y','ID_supplier_y','Period_year_y','Period_month_y','date'],as_index=False).sum()
```

```python
# На графике можно увидеть все просрочки с соответствующими характеристиками
# 'Project_y','Bank','Register_number_y','ID_supplier_y','Period_year_y','Period_month_y','date'
fig = px.scatter(payments_not_in_time, x="date", y="Sum_payment",
                 color="Project_y",
                 size='Sum_payment',
                 hover_data={"date": "|%B %d, %Y","ID_supplier_y":True,"Register_number_y": True,"Bank":True},
                title="Просроченные платежи с разбивкой по полю ID_supplier_y")

fig.update_xaxes(
    dtick="M1",
    tickformat="%b\n%Y",
    ticklabelmode="period")
fig.show()
```

```python
fig.write_html("C:/Users/Владимир/PycharmProjects/chart_by_projects_on_id_supplier_level.html")
```

```python
# Сделаем суммирвание платежей по проектам (суммы по ID_supplier_y будут агрегированные) 
table_for_plot=payments_not_in_time.groupby(['Project_y','Period_year_y','Period_month_y','date'],as_index=False).sum()
```

```python
table_for_plot
```

```python
fig1 = px.scatter(table_for_plot, x="date", 
                 y="Sum_payment",color="Project_y",
                 size='Sum_payment',
                 hover_data={"date": "|%B %d, %Y"},
                title="Просроченные платежи с разбивкой по проектам")
fig1.update_xaxes(
    dtick="M1",
    tickformat="%b\n%Y",
    ticklabelmode="period")
fig1.show()
```

```python
fig1.write_html("C:/Users/Владимир/PycharmProjects/chart_by_projects.html")
```
